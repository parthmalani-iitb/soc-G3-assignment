```python
import yfinance as yf
import pandas as pd
```

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
stock_data = yf.download("RELIANCE.NS", start="2024-06-01", end="2024-07-04")
print(stock_data)
```

```python
nifty_data = yf.download('^NSEI', start="2024-06-01", end="2024-07-04")
print(nifty_data)
```

```python
data_summary = {
    "Data Size": stock_data.shape,
    "Missing Values": stock_data.isnull().sum().sum(),
    "Mean": stock_data.mean(),
    "Median": stock_data.median(),
    "Standard Deviation": stock_data.std()
}
# python Object
```

```python
data_summary
```

```python
# as Missing value is zero so we cannot have to run the function to handle missing values
```

```python
stock_data['Cumulative Return'] = (1 + stock_data['Adj Close'].pct_change()).cumprod()
nifty_data['Cumulative Return'] = (1 + nifty_data['Adj Close'].pct_change()).cumprod()
        
plt.figure(figsize=(10, 6))
plt.plot(stock_data['Cumulative Return'], label=f'RELIANCE.NS Cumulative Return')
plt.plot(nifty_data['Cumulative Return'], label='Nifty Index Cumulative Return')
plt.title(f'RELIANCE.NS vs Nifty Index Cumulative Return')
plt.xlabel('Date')
plt.ylabel('Cumulative Return')
plt.legend()
plt.grid(True)
plt.show()
```
